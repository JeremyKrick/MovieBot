```python
import json
import pandas as pd

# Read your CSV file
data = pd.read_csv('movie_lines.csv').head(50)

# Ensure the data has an even number of rows for pairing. Drop the last row if not.
if len(data) % 2 != 0:
    data = data[:-1]

# Initiate a list to store formatted chat sessions
chat_sessions = []

for i in range(0, len(data), 2):
    # Create a single chat session with user and assistant messages
    session = []
    
    # Optional system instruction
    session.append({
        'role': 'system',
        'content': "You are a chatbot trained on movie lines."
    })
    
    # User's line (from the Character's Text)
    session.append({
        'role': 'user',
        'content': data['Text'].iloc[i]
    })
    
    # Assistant's response (from the next Character's Text)
    session.append({
        'role': 'assistant',
        'content': data['Text'].iloc[i+1]
    })
    
    chat_sessions.append(session)

# Save to JSONL file
with open('chat_formatted_fine-tuning.jsonl', 'w') as outfile:
    for chat in chat_sessions:
        json.dump({'messages': chat}, outfile)
        outfile.write('\n')
```

```python
import json
import pandas as pd

# Read your CSV file
data = pd.read_csv('movie_lines.csv').head(50)

# Ensure the data has an even number of rows for pairing. Drop the last row if not.
if len(data) % 2 != 0:
    data = data[:-1]

# Initiate a list to store formatted prompts and completions
formatted_data = []

for i in range(0, len(data), 2):
    # User's line (from the Character's Text) acts as the prompt
    prompt = data['Text'].iloc[i]
    
    # The next Character's Text acts as the completion
    completion = data['Text'].iloc[i+1]
    
    formatted_data.append({
        'prompt': prompt,
        'completion': completion
    })

# Save to JSONL file
with open('prompt_completion_formatted_fine-tuning.jsonl', 'w') as outfile:
    for item in formatted_data:
        json.dump(item, outfile)
        outfile.write('\n')
```
